```python
import numpy as np
y_pred = np.array([1,1,0,0,1])
y_true = np.array([0.30, 0.7, 1, 0, 0.5])
```

# Mean Absolute Error

image.png

**Without Using Numpy**

```python
# defining a function to calculate MAE
def mae(y_pred, y_true):
  total_error = 0
  for yt, yp in zip(y_pred, y_true):
    total_error += abs(yp - yt)
  print("Total error:", total_error)
  mae = total_error/len(y_true)
  print("Mean Absolute Error:", mae)
  return mae
```

```python
mae(y_pred, y_true)
```

**Implement same thing using numpy in much easier way**

```python
np.abs(y_pred - y_true)
```

```python
np.mean(np.abs(y_pred - y_true))
```

```python
def mae_np(y_pred, y_true):
  return np.mean(np.abs(y_pred, y_true))
```

```python
mae_np(y_pred, y_true)
```

# Mean Squared Error (MSE)

image.png

**Without using Numpy**

```python
import numpy as np
y_pred = np.array([1,1,0,0,1])
y_true = np.array([0.3, 0.7, 1, 0, 0.5])
```

```python
def mse(y_true, y_pred):
  total_error = 0
  for yt, yp in zip(y_true,y_pred):
    total_error += (yt - yp)**2
  print("Total Error:", total_error)
  mse = total_error/len(y_true)
  print("Mean Squared Error:", mse)
  return mse
```

```python
mse(y_true, y_pred)
```

**MSE by using NumPy**

```python
np.mean(np.square(y_true - y_pred))
```

# Log Loss

**Log Loss or Binary Cross Entropy**

```python
np.log([0])
```

```python
np.log([1])
```

```python
def log_loss(y_true, y_pred):
  epsilon = 1e-15
  y_predicted_new = [max(i, epsilon) for i in y_pred]
  y_predicted_new = [min(i, 1-epsilon) for i in y_predicted_new]
  y_predicted_new = np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new) + (1-y_true)*np.log(1-y_predicted_new))
```

```python
log_loss(y_true, y_pred)
```
